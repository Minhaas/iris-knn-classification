```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris['data'][:5])
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd 
import mglearn
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=0)
print(f"X_train shape: {X_train.shape}")
```

```python
iris_df = pd.DataFrame(X_train, columns=iris.feature_names)
print(iris_df[:5])
```

```python
scat = pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(20,20), marker='o', hist_kwds={'bins':30}, cmap=mglearn.cm3, s=60, alpha=8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
import numpy as np
```

```python
X_new = np.array([[5,2.9,3,9]])
prediction = knn.predict(X_new)
print(f"Prediction is: {prediction}")
print(f"Prediction class is: {iris['target_names'][prediction]}")
```

```python
newPred = knn.predict(X_test)
print(f"Test set prediction: {newPred}")
```

```python
accuracy = knn.score(X_test, y_test)
print("Test set accuracy is: {:1f}".format(accuracy))
```
